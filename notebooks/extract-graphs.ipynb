```python
import json
def export_results(nodes, edges, file_name):
    edges = set(edges)
    dict_edges = []
    for edge in edges:
        dict_edges.append({"from": edge[0], "to": edge[1], "relation": edge[2]})
    
    out_json = {
        "nodes": list(set(nodes)),
        "edges": dict_edges
    }

    with open(f"{file_name}.json", "w", encoding="utf-8") as file:
        json.dump(out_json, file)
```

```python
def extract_corpus_graph(corpus_path):

    suffix = "eu"
    
    with open(corpus_path, "r") as f:
        extractions = json.load(f)
    
    nodes = []
    edges = []
    for city_extractions in extractions:
        city = city_extractions["city"]
        nodes.append(city)
        places = city_extractions["extractions"]
        for place_info in places:
            place_name = place_info["extraction_text"]
            place_type = place_info["extraction_class"]

            if place_name == place_type: 
                continue
            
            nodes.append(place_name)
                
            edges.append((place_name, city, "located_in"))
            edges.append((place_name, place_type, "type"))
    
    export_results(nodes, edges, f"corpus_graph_{suffix}")
```

```python
def extract_ontology_graph(ontology_path):

    suffix = "eu"
    
    with open(ontology_path, "r") as f:
        ontology_extractions = json.load(f)
    
    ontology_nodes = []
    ontology_edges = []
    for city, city_info in ontology_extractions.items():
        ontology_nodes.append(city)
        places = city_info["places"]
        for place_info in places:
            place_name = place_info["label"]
            place_type = place_info["type"]

            ontology_nodes.append(place_name)
            ontology_edges.append((place_name, city, "located_in"))
            ontology_edges.append((place_name, place_type, "type"))
    
    export_results(ontology_nodes, ontology_edges, f"ontology_graph_{suffix}")
```
