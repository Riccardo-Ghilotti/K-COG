```python
import json
import os

CORPUS_PATH = "langextract_corpus_extractions_eu"
OUT_FILE_PATH = "merged_corpus_extractions_eu.json"

city_folders = sorted(os.listdir(CORPUS_PATH), key=str.lower)

merged_data = []
for city in city_folders:
    file_path = f"./{CORPUS_PATH}/{city}/out.json"
    with open(file_path, "r", encoding="utf-8") as input_file:
        cur_data = json.load(input_file)
    extractions = cur_data["extractions"]
    city = cur_data["city"]
    city_extractions = {
        "city": city,
        "extractions": extractions
    }
    merged_data.append(city_extractions)
    
    
with open(OUT_FILE_PATH, "w", encoding="utf-8") as output_file:
    json.dump(merged_data, output_file, ensure_ascii=False, indent=4)
```
